```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error
from IPython.display import clear_output, display_html
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
```

```python
data = pd.read_csv('data/zillow_data.csv')
```

```python
def melt_data(df):
    """
    Takes the zillow_data dataset in wide form or a subset of the zillow_dataset.  
    Returns a long-form datetime dataframe 
    with the datetime column names as the index and the values as the 'values' column.
    
    If more than one row is passes in the wide-form dataset, the values column
    will be the mean of the values from the datetime columns in all of the rows.
    """
    
    melted = pd.melt(df, id_vars=['RegionName', 'RegionID', 'SizeRank', 'City', 'State', 'Metro', 'CountyName'], var_name='time')
    melted['time'] = pd.to_datetime(melted['time'], infer_datetime_format=True)
    melted = melted.dropna(subset=['value'])
#     return melted.groupby('time').aggregate({'value':'mean'})
    melted.set_index('time', inplace=True)
    return melted
```

```python
df = melt_data(data)

df['zipcount'] = df['RegionName'].map(df.RegionName.value_counts())

df = df[df.zipcount==265]
```

```python
statistics = []
pvalues = []
cvs = []

for zipcode in df.RegionName.unique():
    
    zc = df[df.RegionName==zipcode].value

    dftest = adfuller(zc)

    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    statistics.append(dfoutput['Test Statistic'])
    pvalues.append(dfoutput['p-value'])
    cvs.append(dfoutput['Critical Value (5%)'])
```

```python
dfdf = pd.DataFrame({'zipcode': df.RegionName.unique(), 'adfuller_statistic': statistics, 'critical_value': cvs, 'adfuller_pvalue': pvalues}).sort_values(by='adfuller_statistic', ascending=False)
```

# Differencing

```python
statistics = []
pvalues = []
cvs = []


for zipcode in dfdf.zipcode:
    zc = df[df.RegionName==zipcode].value
    first_difference = zc - zc.shift(1)
    
    dftest = adfuller(first_difference.iloc[1:])
    
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    statistics.append(dfoutput['Test Statistic'])
    pvalues.append(dfoutput['p-value'])
    cvs.append(dfoutput['Critical Value (5%)'])
    
    
dfdf['fd_statistic'] = statistics
dfdf['fd_critical_value'] = cvs
dfdf['fd_pvalue'] = pvalues
```

```python
statistics = []
pvalues = []
cvs = []


for zipcode in dfdf.zipcode:
    zc = df[df.RegionName==zipcode].value
    first_difference = zc - zc.shift(1)
    seasonal_first_difference = first_difference - first_difference.shift(12)
    
    dftest = adfuller(seasonal_first_difference.iloc[13:])
    
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    statistics.append(dfoutput['Test Statistic'])
    pvalues.append(dfoutput['p-value'])
    cvs.append(dfoutput['Critical Value (5%)'])
    
    
dfdf['sfd_statistic'] = statistics
dfdf['sfd_critical_value'] = cvs
dfdf['sfd_pvalue'] = pvalues
```

```python
mses, rmses, residuals = [], [], []

for i in range(len(dfdf[dfdf.sfd_pvalue<0.05].zipcode)):
    zipcode = dfdf[dfdf.sfd_pvalue<0.05].zipcode.iloc[i]
    prices = df[df.RegionName == zipcode]
    shift = prices.shift(1)
    mse = mean_squared_error(prices.value[1:], shift.value[1:])
    rmse = mean_squared_error(prices.value[1:], shift.value[1:], squared=False)
    residual = shift.value[1:] - prices.value[1:]
    
    mses.append(mse)
    rmses.append(rmse)
    residuals.append(residual)
    
    clear_output(wait=True)
    print(f"Zipcode #{i} done.")
```
